```python
import pandas as pd
import numpy as np


# Define the columns to be read as integers
specific_dtype = {'productId': 'int32'}  # Columns to be read as integers

# Read the CSV file with specified data types
df = pd.read_csv('ZaraMat.csv', dtype=dict(specific_dtype, **{col: 'object' for col in pd.read_csv('ZaraMat.csv', nrows=1).columns if col not in specific_dtype}))
df
```

```python
df.drop(['Unnamed: 0','LINING'], axis=1, inplace=True)
num_columns = df.shape[1]
print("Number of columns:", num_columns)
```

```python
df['origins'] = df['origins'].fillna('Unknown').replace(['', None], 'Unknown')
```

```python
df[df['origins'] == 'Unknown'].count() 
```

```python
# Function to remove columns with only NaN values except for one non-null value
def remove_columns_one_non_null(df):
    columns_to_remove = []
    for col in df.columns:
        non_null_count = df[col].count()
        if non_null_count <= 300:
            columns_to_remove.append(col)
    df = df.drop(columns_to_remove, axis=1)
    return df

# Remove columns with only NaN values and one row with non-null value
filtered_df = remove_columns_one_non_null(df)
filtered_df
```

```python
num_columns = filtered_df.shape[1]
print("Number of columns:", num_columns)
```

```python
# Replace NaN and Null values in the entire DataFrame with '@#'
filtered_df = filtered_df.fillna('@#')

# Replace empty strings in the entire DataFrame with '@#'
filtered_df =filtered_df.replace('', '@#')
```

```python
filtered_df = filtered_df.replace('<br>', '-', regex=True)
```

```python
pattern = r'[^\W\d_]+(?:\s+[^\W\d_]+)*'  # Pattern to extract only words and spaces
filtered_df['Material'] = filtered_df['OUTER SHELL'].str.findall(pattern).apply(lambda x: ' '.join(x))
```

```python
filtered_df['origins'] = filtered_df['origins'].str.replace('Made in', '')
```

```python
filtered_df['Material'] = filtered_df['Material'].fillna('Non-Clothing').replace(['', None], 'Non-Clothing')
```

```python
filtered_df.to_csv('material_list.csv',index=False)
```
